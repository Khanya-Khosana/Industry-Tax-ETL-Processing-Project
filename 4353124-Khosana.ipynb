```python
# Import pandas
import pandas as pd

# 1. EXTRACT

def extract(filepath, batch_size=5):
    batches = pd.read_csv(filepath, chunksize=batch_size)
    return batches

# 2. TRANSFORM

def transform(batches):
    
    results = []
    batch_number = 1

    for batch in batches:

        # 1. average_taxable_income
        batch['average_taxable_income'] = (
            batch['total_taxable_income'] / batch['number_of_firms']
        )

        # 2. average_taxable_income
        batch_average_taxable_income = (
            batch['average_taxable_income'].mean()
        )

        # 3. average_tax_paid
        batch['average_tax_paid'] = (
            batch['total_taxes_paid'] / batch['number_of_firms']
        )

        # 4. average_tax_paid sum
        batch_average_tax_paid = (
            batch['average_tax_paid'].sum()
        )

        # 5. average_cash_tax_paid
        batch['average_cash_tax_paid'] = (
            batch['total_cash_taxes_paid'] / batch['number_of_firms']
        )

        # 6. average_cash_tax_paid sum
        batch_average_cash_tax_paid = (
            batch['average_cash_tax_paid'].sum()
        )

        # Storing batch results
        results.append({
            'batch_number': batch_number,
            'batch_average_taxable_income': batch_average_taxable_income,
            'batch_average_tax_paid': batch_average_tax_paid,
            'batch_average_cash_tax_paid': batch_average_cash_tax_paid
        })

        batch_number += 1

    # Create final dataframe
    final_df = pd.DataFrame(results)

    return final_df


# 3. LOAD
def load(final_df, student_id):

    # a. Print final dataframe
    print(final_df)

    # b. Save dataframe as CSV
    output_file = f"{student_id}-batched_average_taxes.csv"

    final_df.to_csv(output_file, index=False)

    print(f"\nFile saved as: {output_file}")



# File path
filepath = "industry_tax.csv"

# Extract
batches = extract(filepath)

# Transform
final_dataframe = transform(batches)

# Load
load(final_dataframe, "4353124")
```
